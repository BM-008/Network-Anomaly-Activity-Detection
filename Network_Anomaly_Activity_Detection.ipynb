```python
import pandas as pd
```

```python
dataset_train=pd.read_csv('Train_Data.csv')
dataset_test=pd.read_csv('Test_Data.csv')
dataset_train.head(10)
```

```python
dataset_train.shape
```

```python
#dataset_train.isnull().sum()
dataset_test.isnull().sum()
```

```python
dataset_train['attack'].unique()
```

```python
dataset_test.head(10)
```

```python
train_cat=dataset_train.select_dtypes(include="object")
test_cat=dataset_test.select_dtypes(include="object")
test_cat
```

```python
train_num=dataset_train.select_dtypes(include=['float64','int64','float32','int32'])
test_num=dataset_test.select_dtypes(include=['float64','int64','float32','int32'])
```

```python
def outlier_capping(x):
    x = x.clip(upper=x.quantile(0.99))
    x = x.clip(lower=x.quantile(0.01))
    return x

train_num=train_num.apply(outlier_capping)
```

```python
def create_dummies( df, colname ):
    col_dummies = pd.get_dummies(df[colname], prefix=colname, drop_first=True)
    df = pd.concat([df, col_dummies], axis=1)
    df.drop( colname, axis = 1, inplace = True )
    return(df)
```

```python
for c_feature in ['protocoltype', 'service', 'flag', 'attack']:
    train_cat = create_dummies(train_cat,c_feature)
train_cat.head()
```

```python
for c_feature in ['protocoltype', 'service', 'flag']:
    test_cat = create_dummies(test_cat,c_feature)
test_cat.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(sparse_output=False,drop='first')
train_cat=pd.DataFrame(ohe.fit_transform(train_cat),columns=train_cat.columns)
train_cat
```

```python
test_cat=pd.DataFrame(ohe.fit_transform(test_cat),columns=test_cat.columns)
test_cat
```

```python
train_new = pd.concat([train_num, train_cat], axis=1)
test_new = pd.concat([test_num, test_cat], axis=1)
train_new.head()
```

```python
train=train_new
test=test_new
```

```python
train.rename(columns={'attack_normal':'attack'},inplace=True)
```

```python
train_input=train.drop(columns='attack')
train_outout=train['attack']
train_input
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(train_input,train_outout,test_size=0.2,random_state=42)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train1=x_train.drop(columns='service_tftp_u')
x_test1=x_test.drop(columns='service_tftp_u')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train1,y_train)
```

```python
lr.score(x_test1,y_test)*100,lr.score(x_train1,y_train)*100
```

```python
from sklearn.linear_model import RidgeClassifier
```

```python
rc=RidgeClassifier()
rc.fit(x_train1,y_train)
```

```python
rc.score(x_test1,y_test)*100,rc.score(x_train1,y_train)*100
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knc=KNeighborsClassifier(n_neighbors=5)
knc.fit(x_train1,y_train)
```

```python
gnb=GaussianNB()
gnb.fit(x_train1,y_train)
```

```python
gnb.score(x_test1,y_test)*100,gnb.score(x_train1,y_train)*100
```

```python
from sklearn.svm import SVC
```

```python
sv=SVC()
sv.fit(x_train1,y_train)
```

```python
sv.score(x_test1,y_test)*100,sv.score(x_train1,y_train)*100
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
li=[('dc1',DecisionTreeClassifier()),('knc1',KNeighborsClassifier()),('lr1',LogisticRegression())]
```

```python
vc=VotingClassifier(li)
vc.fit(x_train1,y_train)
```

```python
vc.score(x_test1,y_test)*100,vc.score(x_train1,y_train)*100
```
